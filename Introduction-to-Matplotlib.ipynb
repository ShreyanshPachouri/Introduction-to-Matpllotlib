## __Matplotlib example workflow__ ##

```python
# 1.Import Matplotlib and get it ready for plotting on Jupyter
%matplotlib inline 
#This line makes the plot appear in the Jupyter Notebook
import matplotlib.pyplot as plt

#2.Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#3.Setup Plot
fig, ax = plt.subplots(figsize = (10, 5)) #(width, height)

#4.Plot Data
ax.plot(x,y)

#5.Customise plot
ax.set(title="Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

#6. Save and show (Save the whole figure)
fig.savefig("sample-plot.png")
```

```python
#Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(ylabel = "Price");
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
#Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplots option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

#Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

__*When plotting something quickly, use the pyplot method. When plotting something more advanced, use the OO metod.__

```python
#Pyplot method
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100]);
# Sets the limit on the x-axis to start from 45 and end at 100
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#Plot the data
scatter = ax.scatter(x = over_50["age"], y=over_50["chol"],c=over_50["target"]);

#Customise the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize = (10, 10), sharex = True)

#Plot the data
scatter = ax0.scatter(x = over_50["age"], y=over_50["chol"],c=over_50["target"]);

#Customise the plot
ax0.set(title="Heart Disease and Cholesterol Levels", xlabel = "Age", ylabel="Cholesterol");

#Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax0.axhline(over_50["chol"].mean(), linestyle='--');

#Plot the data
scatter = ax1.scatter(x = over_50["age"], y=over_50["thalach"],c=over_50["target"]);

#Customise the plot
ax1.set(title="Heart Disease and max heart rate",xlabel = "Age", ylabel="Max heart rate");

#Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle='--');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
# OO method from scratch
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize = (10, 6))

#Plot the data
scatter = ax.scatter(x = over_50["age"], y=over_50["chol"],c=over_50["target"], cmap="winter");
# cmap changes the color scheme

#Customise the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize = (10, 10), sharex = True)

#Plot the data
scatter = ax0.scatter(x = over_50["age"], y=over_50["chol"],c=over_50["target"], cmap="winter");

#Customise the plot
ax0.set(title="Heart Disease and Cholesterol Levels", xlabel = "Age", ylabel="Cholesterol");

#Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax0.axhline(over_50["chol"].mean(), linestyle='--');
ax0.set_xlim([50, 80])

#Plot the data
scatter = ax1.scatter(x = over_50["age"], y=over_50["thalach"],c=over_50["target"], cmap="winter");

#Customise the plot
ax1.set(title="Heart Disease and max heart rate",xlabel = "Age", ylabel="Max heart rate");
ax1.set_ylim([60, 200])

#Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle='--');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
